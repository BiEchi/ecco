```python
import ecco
```

```python
lm = ecco.from_pretrained('t5-small')
```

```python
# Sanity sake
iids = lm.tokenizer("translate English to German: Prime Minister Narendra Modi and Ministry of Health and Welfare had meetings today to discuss the pandemic", return_tensors="pt")["input_ids"]
iids = lm.to(iids)
lm.tokenizer.batch_decode(lm.model.generate(iids, max_length=1000))
```

```python
def run(text):
    output = lm.generate(text, generate=1000, do_sample=True)
    output.saliency()
    output.saliency(style="detailed")
```

```python
run("translate English to German: Prime Minister Narendra Modi and Ministry of Health and Welfare had meetings today to discuss the pandemic")
```
